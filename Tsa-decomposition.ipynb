```python
import pandas as pd
import statsmodels.tsa.api as tsa
```

```python
df = pd.read_csv('TSLA.csv')
df.head()
```

```python
components = tsa.seasonal_decompose(df['Close'], freq = 4, model = 'additive')
```

```python
ts = (df['Close'].to_frame('Original').assign(Trend = components.trend).assign(Seasonality = components.seasonal).
        assign(Residual = components.resid))
```

```python
ts.plot(subplots = True, figsize = (14, 8))
```
